```python
# Split the 'span' column into 'start_year' and 'end_year'
df[['start_year', 'end_year']] = df['span'].str.split('-', expand=True)

# Convert 'start_year' and 'end_year' to numeric (integer) types
df['start_year'] = pd.to_numeric(df['start_year'])
df['end_year'] = pd.to_numeric(df['end_year'])

# Display the first few rows to check
df.head()
```

```python
#16. What is the average number of runs scored per match for players active between 2000 and 2010?
players_2000_2010=df[(df['start_year']>=2000) & (df['end_year'] <=2010)]
```

```python
runs_per_match=players_2000_2010['runs']/players_2000_2010['matches']
```

```python
average_runs_per_match=runs_per_match.mean()
```

```python
print(average_runs_per_match, "is the average number of runs scored per match for players active between 2000 and 2010")
```

```python
# 17. How has the strike rate of players evolved over different decades (1990s, 2000s, 2010s)?
def classify_decade(year):
    if 1990 <= year < 2000:
        return '1990s'
    elif 2000 <= year < 2010:
        return '2000s'
    elif 2010 <= year < 2020:
        return '2010s'
    else:
        return 'Other'
df['decade'] = df['start_year'].apply(classify_decade)
```

```python
# Calculate the average strike rate for each decade
average_strike_rate_by_decade = df.groupby('decade')['strike_rate'].mean()
print(average_strike_rate_by_decade)
```

```python
#18. Compare the average runs scored by players who started their careers in the 1990s versus those who started in the 2000
players_runs_1990s=df[(df['start_year'] >=1990) & (df['start_year'] < 2000)]
players_runs_2000s=df[(df['start_year'] >=2000) & (df['start_year'] < 2010)]
```

```python
avg_runs_1990s=players_runs_1990s['runs'].mean()
avg_runs_2000s=players_runs_2000s['runs'].mean()
```

```python
print("Average runs scored by players who started in the 1990s:",avg_runs_1990s)
print("Average runs scored by players who started in the 2000s:",avg_runs_2000s)
```
